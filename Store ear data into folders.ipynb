```python
for file in os.listdir():
    if file.split('.')[1] == 'bmp' or file.split('.')[1] == 'jpg':
        folder_name, file_name, _ = file.split('_')
#         print(folder_name, file_name)
        new_path = os.path.join(ROOT_DIR, structure_name, folder_name)
        if os.path.isdir(new_path) == False:
            os.makedirs(new_path)
        shutil.copy(os.path.join(os.getcwd(), file), new_path)
```
